```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sqlalchemy import create_engine

# ConfiguraciÃ³n visual para grÃ¡ficos
sns.set_theme(style="whitegrid")
pd.options.display.float_format = '{:,.2f}'.format

# =============================================================================
# 1. CARGA DE DATOS MAESTRA
# =============================================================================
print("ðŸš€ Iniciando AnÃ¡lisis de Business Intelligence...")

# ConexiÃ³n DB
DB_USER = 'analista_medhos'
DB_PASS = 'Medhos2025!'
DB_HOST = 'postgres'
DB_NAME = 'medhos_dw'
engine = create_engine(f'postgresql+psycopg2://{DB_USER}:{DB_PASS}@{DB_HOST}:5432/{DB_NAME}')

# A) Cargar Movimientos (Solo Entradas y Salidas reales)
print("   ðŸ“¥ Leyendo Movimientos desde SIGA...")
query_mov = """
SELECT descripcion_insumo, cod_insumo, fecha_movimiento, 
       tipo_archivo_detectado, cantidad, precio_total, nombre_archivo, 
       deposito_origen_cod, deposito_destino_desc
FROM raw_movimientos_siga 
WHERE tipo_archivo_detectado IN ('ENTRADA', 'SALIDA')
"""
df = pd.read_sql(query_mov, engine)

# B) Cargar DÃ³lar
print("   ðŸ“¥ Leyendo Cotizaciones HistÃ³ricas...")
query_dolar = "SELECT fecha, tipo_cambio_oficial_venta FROM dim_cotizaciones"
df_dolar = pd.read_sql(query_dolar, engine)

# =============================================================================
# 2. PROCESAMIENTO Y DOLARIZACIÃ“N
# =============================================================================
# Asegurar tipos de dato datetime para el cruce
df['fecha_movimiento'] = pd.to_datetime(df['fecha_movimiento'])
df_dolar['fecha'] = pd.to_datetime(df_dolar['fecha'])

# Ordenar para merge_asof (requerido por pandas)
df = df.sort_values('fecha_movimiento')
df_dolar = df_dolar.sort_values('fecha')

# CRUCE INTELIGENTE: Busca la cotizaciÃ³n mÃ¡s cercana anterior (backward)
df = pd.merge_asof(df, df_dolar, left_on='fecha_movimiento', right_on='fecha', direction='backward')

# CÃ¡lculo Financiero
df['cotizacion_usd'] = df['tipo_cambio_oficial_venta'].ffill().bfill() # Manejo de nulos extremos
df['precio_total_usd'] = df['precio_total'] / df['cotizacion_usd']
df['anio'] = df['fecha_movimiento'].dt.year

# ClasificaciÃ³n de Rubro (Para SegmentaciÃ³n posterior)
mapa_rubros = {'2311': 'MEDICAMENTOS', '2678': 'BIOMEDICOS', '2878': 'LABORATORIO', '2015': 'ODONTOLOGIA'}
df['rubro'] = df['deposito_origen_cod'].map(mapa_rubros).fillna('OTROS')

print(f"âœ… Datos Consolidados: {len(df):,} registros.")
print(f"ðŸ’° Gasto HistÃ³rico Total (Entradas): USD {df[df['tipo_archivo_detectado']=='ENTRADA']['precio_total_usd'].sum():,.2f}")

# =============================================================================
# 3. REPORTE DE CALIDAD (EDA)
# =============================================================================
print("\nðŸ“‹ --- REPORTE DE CALIDAD ---")
nulos = df[['precio_total', 'cotizacion_usd']].isnull().sum()
print(f"Registros sin cotizaciÃ³n: {nulos['cotizacion_usd']}")
print(f"Registros con precio 0: {len(df[df['precio_total'] == 0])}")

# =============================================================================
# 4. TABLA PIVOT 1: GASTO FINANCIERO (USD)
# =============================================================================
print("\nðŸ’µ --- EVOLUCIÃ“N DEL GASTO POR INSUMO (Top 15 USD) ---")
# Filtramos SALIDAS (Consumo de hospitales)
df_salidas = df[df['tipo_archivo_detectado'] == 'SALIDA'].copy()

pivot_usd = df_salidas.pivot_table(
    index=['descripcion_insumo', 'cod_insumo'], columns='anio', values='precio_total_usd', aggfunc='sum', fill_value=0
)
pivot_usd['Total_Historico_USD'] = pivot_usd.sum(axis=1)
pivot_usd = pivot_usd.sort_values('Total_Historico_USD', ascending=False)

# Formatear columnas
pivot_usd.columns = [f"{c} (USD)" if isinstance(c, int) else c for c in pivot_usd.columns]
display(pivot_usd.head(15))

# =============================================================================
# 5. TABLA PIVOT 2: CONSUMO FÃSICO (UNIDADES)
# =============================================================================
print("\nðŸ“¦ --- EVOLUCIÃ“N DEL VOLUMEN POR INSUMO (Top 15 Unidades) ---")
pivot_cant = df_salidas.pivot_table(
    index=['descripcion_insumo', 'cod_insumo'], columns='anio', values='cantidad', aggfunc='sum', fill_value=0
)
pivot_cant['Total_Unidades'] = pivot_cant.sum(axis=1)
pivot_cant = pivot_cant.sort_values('Total_Unidades', ascending=False)

# Formatear columnas
pivot_cant.columns = [f"{c} (U.)" if isinstance(c, int) else c for c in pivot_cant.columns]
display(pivot_cant.head(15))

# =============================================================================
# 6. VISUALIZACIÃ“N DE TENDENCIA
# =============================================================================
print("\nðŸ“ˆ --- TENDENCIA DE GASTO REAL MENSUAL ---")
df_salidas['periodo'] = df_salidas['fecha_movimiento'].dt.to_period('M')
gasto_mensual = df_salidas.groupby('periodo')['precio_total_usd'].sum()

plt.figure(figsize=(14, 6))
gasto_mensual.plot(linewidth=2.5, color='#2c3e50', marker='o', markersize=4)
plt.title('EvoluciÃ³n del Gasto Mensual Real (Ajustado por DÃ³lar Oficial)', fontsize=14)
plt.ylabel('Monto en USD')
plt.xlabel('Periodo')
plt.grid(True, alpha=0.3)
plt.show()
```
